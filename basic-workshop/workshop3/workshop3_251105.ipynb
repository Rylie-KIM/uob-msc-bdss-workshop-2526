```python
# class attendance

monday = {"Alice", "Bob", "Charlie", "David", "Emma"}
tuesday = {"Bob", "Charlie", "Emma", "Frank", "Grace"}

student_set = set()
monday_students_set = set()
tuesday_students_set = set()
both_students_set = set()

for student in monday:
    student_set.add(student)
for student in tuesday:
    student_set.add(student)

for student in student_set:
    if student not in monday:
        monday_students_set.add(student)
    if student not in tuesday:
        tuesday_students_set.add(student)
    else:
        both_students_set.add(student)

print(monday_students_set)
print(tuesday_students_set)
print(both_students_set)
print("Alice" in monday_students_set)
tuesday_students_set.add("Hannah")
print(tuesday_students_set)
```

```python
# üéØ Let's count how many times each character is repeated.
sentence = "The quick brown fox jumped over the lazy dog"
word_count = dict()
for character in list(sentence.lower()):
    if character not in word_count:
        word_count[character] = 1
    else:
        word_count[character] += 1
print(word_count)

#sol2
for character in set(sentence.lower()):
    word_count[character] = word_count.get(character, 0) + 1
```

```python
from typing import Optional

# exception handling
try:
    age = int(input("How old are you? "))
    print(f"You are {age} years old.")

except ValueError:
    print("Wrong input.")

else:
    age = age * -1
    print(f"No error occurred.")

finally: # regardless of error, the code block is ran. We usually use for cleaning up the system.
    print("Goodbye!")

```
